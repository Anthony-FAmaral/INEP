```python
# Analise das colunas para entender a problemática do DataFrame e suas variaveis.
dados = pd.read_excel('divulgacao_regioes_ufs_ideb_2023.xlsx')
print(dados.head())
```

```python
#Identificando as linhas que vão ser excluidas e a que se tornará cabeçalho
pd.set_option('display.max_columns', None)
dados
```

```python
#Retirando as 8 primeiras linhas da planilha
dados = pd.read_excel('divulgacao_regioes_ufs_ideb_2023.xlsx', 
                     skiprows=8, 
                     header=0)  # header=0 usa a próxima linha como cabeçalho
```

```python
print(dados.head()) 
print(dados.columns) 
```

```python
# Nomeando as variaves qualitativas do cabeçalho
dados = dados.rename(columns={
    'Unnamed: 0': 'Região',
    'Unnamed: 1': 'Unidade_Federação',
    '1º ao 5º ano': 'Rede'
})
print(dados.columns.tolist())
```

### Selecionando as variaveis cruciais para minha analise

```python
#Essas colunas são exatamente que trazem os valores anuais, UF e tipo de ensino para gerar os insights
colunas_selecionadas = ['Região', 'Unidade_Federação', 2007, 2009, 2011, 2013, 2015, 2017, 2019, 2021]

dados_etl = dados[colunas_selecionadas].copy()
print(dados_etl.head())
```

```python
# Contando os valores nulos existem em cada coluna
print(dados_etl.isnull().sum())
```

```python
print(dados_etl.head(200))
```

```python
# Remove linhas que começam com 'Total' ou que sejam NaN na coluna 'Unidade_Federação'
dados_etl = dados_etl[~dados_etl['Unidade_Federação'].str.startswith('Total', na=False)]
dados_etl = dados_etl.dropna(subset=['Unidade_Federação'])  # Remove as linhas com NaN na coluna 'Unidade_Federação'

# Substitui os valores específicos na coluna 'Unidade_Federação'
dados_etl['Unidade_Federação'] = dados_etl['Unidade_Federação'].replace({
    'Privada (2)': 'Privada',
    'Privada (1)': 'Privada',
    'Pública (4)': 'Pública'
})

# Exibe o DataFrame resultante
print(dados_etl)

```

```python
print("Colunas disponíveis no DataFrame:")
print(dados_etl.columns.tolist())
```

### Modelo Tabular

```python
#A importancia de tranformar linhas em colunas é ecessial para que as medias funcione no power bi com precisão
# Tranformando planilha horizontal em planilha vertical, ou seja colunar.
# Transformação das colunas de ano em linha e uma nova coluna de metas.

dados_transformados = pd.melt(
    dados_etl,
    id_vars=['Região', 'Unidade_Federação'], 
    value_vars=[2007, 2009, 2011, 2013, 2015, 2017, 2019, 2021], 
    var_name='Ano',               
    value_name='Metas'                        
)

# Converter ano para inteiro
dados_transformados['Ano'] = dados_transformados['Ano'].astype(int)
dados_transformados['Metas'] = dados_transformados['Metas'].astype(float)

# Remover linhas com valores nulos em Metas
print (dados_transformados)

```

## Analise Estatíca

```python
# Numero de linha; Valores nulos; Tipagem das variaveis
print(dados_transformados.info())
print(dados_transformados.describe())
```

```python
#Agregações
dados_transformados.groupby('Região')['Metas'].agg([
    'mean',   # média
    'count',  # contagem
    'max',    # valor máximo
    'min'     # valor mínimo
])

```

```python
dados_transformados.to_excel('metas_educacionais_transformadas.xlsx', index=False, sheet_name='Metas')
```
